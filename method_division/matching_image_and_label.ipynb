```python
import os
import glob
from PIL import Image
import json
import shutil
```

```python
# 설정값
train_src_path = "C:\\Users\\herji\\Downloads\\168.한국 전통 수묵화 화풍별 제작 데이터\\01.데이터\\Training\\01.원천데이터\\TS_1" # train_dir
#test_src_path="C:\\Users\\herji\\Downloads\\168.한국 전통 수묵화 화풍별 제작 데이터\\01.데이터\\Validation\\01.원천데이터\\VS_1" # test_dir
label_path = "C:\\Users\\herji\\Downloads\\168.한국 전통 수묵화 화풍별 제작 데이터\\01.데이터\\Training\\02.라벨링데이터\\TL_1" # label_dir
```

```python

train_image_files = glob.glob(os.path.join(train_src_path, "*.jpg"))
#test_image_files=glob.glob(os.path.join(test_src_path,"*.jpg"))    # 현재 validation data의 라벨링 데이터가 존재하지 않음 
image_files=train_image_files#+test_image_files
label_files = glob.glob(os.path.join(label_path, "*.json"))
```

```python
data_list = []

for image_file in image_files:

    image = Image.open(image_file)
    

    image_filename = os.path.splitext(os.path.basename(image_file))[0]  # 확장자 제외한 파일 이름
    matching_label_file = next((label for label in label_files if image_filename in label), None) # 해당 이미지 파일의 라벨 파일 찾기 (파일 이름을 매칭,맞는 것이 없으면 NONE)
    
    if matching_label_file:

        with open(matching_label_file, 'r',encoding='utf-8') as label_json: #label_json이라는 이름으로 라벨링 파일 읽기
            label_data = json.load(label_json) 
        
       
        data_list.append({'image': image, 'label': label_data})  # 이미지와 라벨 데이터 리스트 안에 딕셔너리가 들어 있는 형태로 저장
    else:
        print(f"No label file found for {image_filename}")  # 이미지와 라벨데이터가 매칭되지 않으면 오류 출력
    image.close()


```
